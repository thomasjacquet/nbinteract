```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

chk = widgets.ToggleButton(value=False,description="Empezar Simulacion",tooltip="Description",icon="check-circle")
chkbox_1 = widgets.RadioButtons(options=["TARJETA DEBITO","TARJETA CREDITO"],value = "TARJETA DEBITO",description="Tarjeta :")
chkbox_2 = widgets.RadioButtons(options=["Física","Jurídica"],value = "Física",description="Persona :")
chkbox_3 = widgets.RadioButtons(options=["Si","No"],value = "No",description="Autónomo :")
btb = widgets.BoundedFloatText(value=0.0,min=0,max=100000000,description="Importe :")

@interact(apply=widgets.Checkbox(value=False,description="Aplicar nuevas tarifas"),tarjeta=widgets.RadioButtons(options=["TARJETA DEBITO","TARJETA CREDITO"],value = "TARJETA DEBITO",description="Tarjeta :"), 
          persona=widgets.RadioButtons(options=["Física","Jurídica"],value = "Física",description="Persona :"),
          autonomo= widgets.RadioButtons(options=["Si","No"],value = "No",description="Autónomo :"),importe = widgets.BoundedFloatText(value=0.0,min=0,max=100000000,description="Importe :"))
def simulador(apply,tarjeta,persona,autonomo,importe ):
  while True:
    if apply == False:   
      if tarjeta == 'TARJETA DEBITO':                        
        if importe <= 20:                           
          comision = importe*0.001                  
          if comision >= 0.07:                                    
            comision = 0.07                                                                                                              
        elif importe  > 20:                        
          comision = importe *0.002            
          if comision >= 0.07:
            comision = 0.07
      elif tarjeta == 'TARJETA CREDITO':
        if importe <= 20: 
          comision = importe*0.002
          if comision >= 0.07:
            comision = 0.07
        elif importe > 20:
          comision = importe*0.003
          if comision >= 0.07:
            comision = 0.07
      return comision
    else:
      if persona == 'Física' and autonomo == 'No': 
          if tarjeta == 'TARJETA DEBITO':                        
            if importe <= 20:                           
              comision = importe*0.001                  
              if comision >= 0.07:                                    
                comision = 0.07                                                                              
            elif importe > 20:                        
              comision = importe*0.002            
              if comision >= 0.07:
                comision = 0.07
          elif tarjeta == 'TARJETA CREDITO':
            if importe <= 20: 
              comision = importe*0.002
              if comision >= 0.07:
                comision = 0.07
            elif importe > 20:
              comision = importe*0.003
              if comision >= 0.07:
                comision = 0.07
      elif persona == 'Jurídica' or autonomo == 'Si':
        if tarjeta == 'TARJETA DEBITO':
          comision = 0.35 
        elif tarjeta == 'TARJETA CREDITO':
          comision = importe*0.0076
      return comision
```
